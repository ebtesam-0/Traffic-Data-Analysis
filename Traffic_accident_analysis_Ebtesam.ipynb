```python
# هنا نحسب عدد القيم المفقودة (NaN) في كل عمود
missing_data = accident_df.isnull().sum()  # نستخدم isnull() عشان نحدد القيم المفقودة (NaN) في كل عمود، وبعدين نستخدم sum() عشان نحسب عدد القيم المفقودة في كل عمود

# هنا نجيب نوع البيانات (رقم، نص، تاريخ...) لكل عمود
data_types = accident_df.dtypes  # نستخدم dtypes عشان نجيب نوع البيانات لكل عمود في الجدول (مثل: رقم، نص، تاريخ)

# ندمج النتايج في DataFrame واحد فيه عمودين:
# Missing Values = عدد القيم المفقودة
# Data Type = نوع البيانات
missing_info = pd.DataFrame({'Missing Values': missing_data, 'Data Type': data_types})  # ننشئ DataFrame جديد فيه العمودين: القيم المفقودة ونوع البيانات

# نعرض فقط الأعمدة اللي فيها قيم مفقودة (Missing Values > 0)
missing_info[missing_info['Missing Values'] > 0]  # نعرض الأعمدة اللي تحتوي على قيم مفقودة فقط عن طريق تصفيتها بناءً على القيم المفقودة
```

```python
# نحدد الأعمدة الرقمية (float64) والنصية (object)
numeric_cols = accident_df.select_dtypes(include=['float64']).columns  # نحدد الأعمدة اللي نوع بياناتها رقمي (float64)
text_cols = accident_df.select_dtypes(include=['object']).columns  # نحدد الأعمدة اللي نوع بياناتها نصي (object)

# للأعمدة الرقمية
# السبب: استخدمنا "median" (الوسيط) بدل mean
# لأن الوسيط ما يتأثر بالقيم الشاذة (outliers) اللي ممكن تلخبط المتوسط
accident_df[numeric_cols] = accident_df[numeric_cols].fillna(accident_df[numeric_cols].median())  # نعوض القيم المفقودة في الأعمدة الرقمية باستخدام الوسيط (median)

# للأعمدة النصية
# السبب: استخدمنا "mode" (الأكثر تكرار)
# لأن النصوص غالباً فيها قيم متكررة بشكل كبير
# وتعويض القيم الناقصة بالأكثر تكرار يعطي نتيجة أقرب للواقع
for col in text_cols:  # نكرر العملية على كل عمود نصي
    accident_df[col] = accident_df[col].fillna(accident_df[col].mode()[0])  # نعوض القيم المفقودة بالقيمة الأكثر تكرار في العمود
```

```python
# ننشئ دالة (function) اسمها time_of_day
# الفايدة: نصنّف الساعة إلى فترة زمنية (صباح / ظهر / مساء / ليل)
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'   # صباح (من 5 الفجر إلى 11:59)
    elif 12 <= hour < 17:
        return 'Afternoon' # ظهر (من 12 الظهر إلى 4:59 العصر)
    elif 17 <= hour < 21:
        return 'Evening'   # مساء (من 5 العصر إلى 8:59 الليل)
    else:
        return 'Night'     # ليل (من 9 الليل إلى 4:59 الفجر)

# نطبق الدالة على عمود الساعة (Hour)
Fdata['Time_of_Day'] = Fdata['Hour'].apply(time_of_day)  # نضيف عمود جديد لتحديد الفترة الزمنية لكل ساعة

# نعرض أول 5 صفوف من الجدول مع الأعمدة المهمة
print(Fdata[['Date', 'Hour', 'Time_of_Day']].head())  # نعرض الأعمدة المهمة: التاريخ، الساعة، والفترة الزمنية
```

```python
# ننشئ دالة لتبسيط حالة الطقس
# الفايدة: نوحّد التصنيفات بدل ما تكون كثيرة وصعبة (مثلاً: "heavy rain" و "light rain" كلهم نصنّفهم "Rain")
def simplify_weather_condition(condition):
    condition = str(condition).lower()  # نخلي النص كله حروف صغيرة عشان نسهل المقارنة
    if 'rain' in condition or 'shower' in condition:
        return 'Rain'   # مطر
    elif 'fog' in condition:
        return 'Fog'    # ضباب
    elif 'snow' in condition:
        return 'Snow'   # ثلج
    elif 'fine' in condition or 'sun' in condition:  # نضيف "fine" كشرط للمشمس
        return 'Clear'  # مشمس / صافٍ
    else:
        return 'Other'  # غير مصنّف (أي حالة طقس ثانية)

# نطبق الدالة على عمود الطقس ونخزن النتيجة بعمود جديد
accident_df['Simplified_Weather_Conditions'] = accident_df['Weather_Conditions'].apply(simplify_weather_condition)

# نعرض أول 5 صفوف بس مع الأعمدة المهمة
print(accident_df[['Date', 'Weather_Conditions', 'Simplified_Weather_Conditions']].head())  # نعرض الأعمدة المهمة: التاريخ، حالة الطقس الأصلية، وحالة الطقس المبسطة
```

```python
# نتأكد أن عمود عدد المركبات ما فيه قيم مفقودة
# - fillna(0): إذا فيه قيمة ناقصة (NaN) نحط بدلها صفر
# - astype(int): نحول العمود كله لأرقام صحيحة (integer)
accident_df['Number_of_Vehicles'] = accident_df['Number_of_Vehicles'].fillna(0).astype(int)

# نعرض أول 5 صفوف من الجدول مع التاريخ وعدد المركبات
print(accident_df[['Date', 'Number_of_Vehicles']].head())  # نعرض الأعمدة المهمة: التاريخ وعدد المركبات
```

```python
# هنا نسوي قاموس (Dictionary) نحدد فيه التصنيف
# الفكرة: نربط كل قيمة من العمود الأصلي بقيمة جديدة أبسط
severity_mapping = {
    'Fatal': 'Fatal',      # حاد (حادث مميت)
    'Serious': 'Serious',  # خطير (إصابات قوية)
    'Slight': 'Slight'     # طفيف (إصابات بسيطة)
}

# نضيف عمود جديد فيه التصنيف المبسط
# الفايدة: نستخدم map مع القاموس اللي فوق عشان نحول القيم في عمود "Accident_Severity" إلى قيم مبسطة وسهلة الفهم
accident_df['Accident_Severity_Classified'] = accident_df['Accident_Severity'].map(severity_mapping)

# نعرض أول 5 صفوف بس للتأكد أن العمود الجديد انضاف بشكل صحيح
print(accident_df[['Accident_Severity', 'Accident_Severity_Classified']].head())  # نعرض العمود الأصلي والعمود الجديد للتأكد من صحة التحويل
```

```python
# نعيد تحميل البيانات من الملف (عشان نتأكد أن كل الأعمدة موجودة)
data_path = path  # نستخدم المتغير path اللي يحتوي على المسار الصحيح للبيانات
accident_df = pd.read_csv(data_path + "/Accident_Information.csv", low_memory=False, encoding='latin-1')

# دالة تبسيط حالة الطقس
# الفايدة: نوحّد التصنيفات بدل ما تكون كثيرة (مثلاً "heavy rain" و "light rain" كلهم نصنّفهم "Rain")
def simplify_weather_condition(condition):
    condition = str(condition).lower()  # نحول النص لحروف صغيرة عشان نقارن بسهولة
    if 'rain' in condition or 'shower' in condition:
        return 'Rain'   # مطر
    elif 'fog' in condition:
        return 'Fog'    # ضباب
    elif 'snow' in condition:
        return 'Snow'   # ثلج
    elif 'fine' in condition or 'sun' in condition:
        return 'Clear'  # مشمس / صافٍ
    else:
        return 'Other'  # أي حالة ثانية

# نطبق الدالة على عمود الطقس ونخزن النتيجة بعمود جديد
accident_df['Simplified_Weather_Conditions'] = accident_df['Weather_Conditions'].apply(simplify_weather_condition)
# نضيف عمود جديد في البيانات يحتوي على حالة الطقس المبسطة

# نرسم تأثير الطقس على عدد الحوادث
plt.figure(figsize=(8, 6))  # نحدد حجم الرسم (عرض 8 × ارتفاع 6)

sns.countplot(x='Simplified_Weather_Conditions', data=accident_df)
# countplot: يرسم أعمدة بعدد الحوادث في كل حالة طقس (Rain, Fog, Snow, Clear, Other)

plt.title('Accidents by Weather Conditions')
# عنوان الرسم يوضح توزيع الحوادث حسب حالة الطقس

plt.xlabel('Weather Conditions')
# تسمية المحور الأفقي (حالة الطقس)

plt.ylabel('Number of Accidents')
# تسمية المحور العمودي (عدد الحوادث)

plt.xticks(rotation=45)
# نلف أسماء الأعمدة شوي (45 درجة) عشان ما تتداخل وتكون واضحة

plt.show()  # نعرض الرسم البياني
```

```python
# نحسب عدد الحوادث لكل منطقة / مدينة
accidents_by_area = accident_df['Local_Authority_(District)'].value_counts()
# نستخدم value_counts() لحساب عدد الحوادث في كل منطقة أو مدينة

# نأخذ أعلى 10 مناطق فقط (اللي فيها أكبر عدد حوادث)
top_areas = accidents_by_area.head(10).index
# نختار أعلى 10 مناطق تحتوي على أكبر عدد من الحوادث

# نصفي البيانات بحيث تحتوي فقط على هذي المناطق العشرة
accident_df_top_areas = accident_df[accident_df['Local_Authority_(District)'].isin(top_areas)]
# نصفي البيانات لتشمل فقط الحوادث في أعلى 10 مناطق

# نرسم توزيع الحوادث حسب المدن/المناطق (مع عرض أعلى 10 فقط)
plt.figure(figsize=(10, 6))  # نحدد حجم الشكل (عرض 10 × ارتفاع 6)

sns.countplot(x='Local_Authority_(District)', data=accident_df_top_areas)
# نرسم الأعمدة بحيث كل عمود يمثل منطقة وعدد الحوادث فيها

plt.title('Top 10 Accidents by Local Authority (District)')
# عنوان الرسم: توزيع الحوادث حسب المناطق العشرة الأعلى

plt.xlabel('Local Authority (District)')
# تسمية المحور الأفقي (أسماء المدن/المناطق)

plt.ylabel('Number of Accidents')
# تسمية المحور العمودي (عدد الحوادث)

plt.xticks(rotation=90)
# نلف أسماء المدن 90 درجة عشان تقرأها بسهولة

plt.show()  # نعرض الرسم البياني
```

```python
# نتأكد أن عمود التاريخ بصيغة datetime
accident_df['Date'] = pd.to_datetime(accident_df['Date'], errors='coerce')
# الفايدة: نحول عمود التاريخ إلى صيغة datetime عشان نقدر نستخدمه في العمليات الحسابية واستخراج التفاصيل

# نضيف أعمدة جديدة: الشهر + يوم الأسبوع
accident_df['Month'] = accident_df['Date'].dt.month  # رقم الشهر (1-12)
accident_df['Day_of_Week'] = accident_df['Date'].dt.dayofweek  # يوم الأسبوع (0=الاثنين ... 6=الأحد)

# نحسب عدد الحوادث حسب (الشهر × يوم الأسبوع)
# groupby: نجمع البيانات على أساس شهر وأيام الأسبوع
# size(): نعد عدد الحوادث
# unstack(): نحول النتيجة إلى جدول (rows = شهور، columns = أيام الأسبوع)
heatmap_data = accident_df.groupby(['Month', 'Day_of_Week']).size().unstack()

# نرسم الخريطة الحرارية
plt.figure(figsize=(8, 6))  # نحدد حجم الشكل (عرض 8 × ارتفاع 6)

sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu')
# annot=True: يكتب الأرقام داخل المربعات
# fmt='d': يخلي الأرقام بصيغة عدد صحيح
# cmap='YlGnBu': يحدد ألوان الخريطة (من أصفر إلى أزرق)

plt.title('Heatmap of Accidents by Month and Day of the Week')
# عنوان الرسم يوضح توزيع الحوادث حسب الشهر ويوم الأسبوع

plt.xlabel('Day of the Week')
# المحور الأفقي = أيام الأسبوع

plt.ylabel('Month')
# المحور العمودي = الشهور

plt.show()  # نعرض الخريطة الحرارية
```

```python
# هذا الرسم يوضح خريطة حرارية (Heatmap) لعدد الحوادث
# محور Y يمثل الأشهر (من 1 إلى 12)
# محور X يمثل أيام الأسبوع (0 إلى 6)
# - غالباً 0 = الأحد أو الاثنين حسب طريقة ترميز البيانات
#
# الألوان توضح كثافة وعدد الحوادث:
# - كل ما كان اللون أزرق غامق أكثر = عدد الحوادث أكبر
# - كل ما كان اللون فاتح يميل للأصفر = عدد الحوادث أقل
#
# من الملاحظ:
# - الحوادث تكثر في منتصف الأسبوع (الأيام 2 و3) خصوصاً في الشهور 10 و11
# - تقل الحوادث في نهاية الأسبوع (اليوم 6)
# - في بعض الأشهر (زي 7 و8 و11) الحوادث أعلى بشكل عام مقارنة بالأشهر الثانية
#
# الخريطة تساعدنا نعرف العلاقة بين (الشهر + يوم الأسبوع) وعدد الحوادث،
# وبالتالي نقدر نحدد الأوقات اللي تحتاج إجراءات سلامة أو مراقبة أكثر.
```

```python
# نعيد إنشاء عمود "Accident_Severity_Classified" باستخدام القاموس
severity_mapping = {
    'Fatal': 'Fatal',      # حادث حاد (مميت)
    'Serious': 'Serious',  # حادث خطير
    'Slight': 'Slight'     # حادث طفيف
}
accident_df['Accident_Severity_Classified'] = accident_df['Accident_Severity'].map(severity_mapping)

# نحسب عدد كل نوع من الحوادث
accident_types = accident_df['Accident_Severity_Classified'].value_counts()

# نرسم رسم دائري (Pie chart) يوضح توزيع أنواع الحوادث
plt.figure(figsize=(8, 6))  # نحدد حجم الشكل (عرض 8 × ارتفاع 6)

accident_types.plot.pie(
    autopct='%1.1f%%',             # يكتب النسبة المئوية داخل كل جزء
    startangle=90,                 # يبدأ الرسم من الأعلى (زاوية 90)
    colors=['#A8DADC', '#457B9D', '#F1FAEE']  # ألوان رايقة: أزرق فاتح، أزرق داكن، أبيض مائل للرمادي
)

plt.title('Distribution of Accidents by Severity')
# عنوان الرسم

plt.ylabel('')
# نخلي التسمية فاضية عشان الشكل يكون أنظف

plt.show()  # نعرض الرسم البياني
```

```python
# أول شي نحول درجة الخطورة إلى أرقام عشان نقدر نحسب المتوسط
severity_numeric = {
    'Slight': 1,   # حادث طفيف
    'Serious': 2,  # حادث خطير
    'Fatal': 3     # حادث مميت
}
accident_df['Severity_Numeric'] = accident_df['Accident_Severity'].map(severity_numeric)
# نضيف عمود جديد "Severity_Numeric" يحتوي على القيم الرقمية المقابلة لكل درجة خطورة

# نرسم مقارنة لمتوسط الخطورة حسب حالة الطقس
plt.figure(figsize=(8, 6))  # نحدد حجم الشكل (عرض 8 × ارتفاع 6)

sns.barplot(
    x='Simplified_Weather_Conditions',  # المحور الأفقي: حالة الطقس
    y='Severity_Numeric',               # المحور العمودي: درجة الخطورة الرقمية
    data=accident_df,                   # نستخدم البيانات من accident_df
    estimator='mean',                   # نحسب المتوسط لكل حالة طقس
    palette=['#4B9CD3', '#2D79B9', '#1A5F84', '#A0C9E6']  # ألوان بحرية هادئة: أزرق فاتح، أزرق داكن، أزرق بحري
)

plt.title('Average Severity by Weather Conditions')
# عنوان الرسم يوضح المتوسط لدرجة الخطورة حسب حالة الطقس

plt.xlabel('Weather Conditions')
# المحور الأفقي = حالة الطقس

plt.ylabel('Average Accident Severity')
# المحور العمودي = متوسط درجة الخطورة (1=طفيف, 2=خطير, 3=حاد)

plt.xticks(rotation=45)
# نلف أسماء الحالات شوي لتكون واضحة

plt.show()  # نعرض الرسم البياني
```

```python
# هذا الرسم يوضح متوسط خطورة الحوادث (Average Accident Severity)
# حسب حالة الطقس (Weather Conditions).
#
# المحور الأفقي (X): أنواع الطقس (مطر، صافٍ، ثلج، ضباب، إلخ).
# المحور العمودي (Y): متوسط درجة خطورة الحوادث.
#
# الملاحظات:
# - الضباب (Fog) يبين أعلى متوسط خطورة بين باقي الحالات.
# - الحوادث في الأجواء الصافية (Clear) تجي بعده تقريباً.
# - المطر (Rain) قريب منها لكن أقل شوي.
# - الثلج (Snow) وحالات أخرى (Other) يظهرون بأقل خطورة نسبياً.
#
# الفايدة من الرسم:
# يساعدنا نعرف كيف حالة الطقس تأثر على شدة الحوادث،
# وبالتالي نركز على إجراءات السلامة في الظروف اللي تزيد فيها الخطورة (زي الضباب).
```

```python
# نحسب عدد الحوادث لكل سنة
accidents_by_year = accident_df['Year'].value_counts().sort_index()
# نستخدم value_counts() لحساب عدد الحوادث في كل سنة، ثم نرتبها حسب السنة باستخدام sort_index()

plt.figure(figsize=(8, 6))  # نحدد حجم الشكل (عرض 8 × ارتفاع 6)

# نرسم خط بلون أزرق مع دوائر على كل سنة
accidents_by_year.plot(
    kind='line',       # نوع الرسم خط (Line)
    color='purple',      # لون الخط (أزرق)
    marker='o',        # شكل النقطة (o = دائرة)
    linestyle='-',     # نوع الخط (مستقيم)
    linewidth=2,       # سمك الخط
    markersize=6       # حجم النقاط
)

plt.title('Trend of Accidents Over Years')
# عنوان الرسم يوضح الاتجاه الزمني لعدد الحوادث

plt.xlabel('Year')
# تسمية المحور الأفقي (السنة)

plt.ylabel('Number of Accidents')
# تسمية المحور العمودي (عدد الحوادث)

plt.grid(True, linestyle='--', alpha=0.6)  # نضيف شبكة خفيفة للتوضيح (خطوط متقطعة بدرجة شفافية 0.6)

plt.show()  # نعرض الرسم البياني
```

```python
# هذا الرسم يوضح اتجاه الحوادث (Trend of Accidents) عبر السنوات من 2005 إلى 2017.
#
# المحور الأفقي (X): يمثل السنوات.
# المحور العمودي (Y): عدد الحوادث في كل سنة.
#
# الملاحظات:
# - واضح إن عدد الحوادث كان مرتفع جداً في 2005 (قريب من 200 ألف حادث).
# - بعدها بدأ ينخفض تدريجياً سنة بعد سنة.
# - في 2013 كان فيه أقل مستوى تقريباً (حوالي 139 ألف حادث).
# - في 2014 حصل ارتفاع بسيط (رجع إلى حوالي 147 ألف).
# - بعدها رجع ينخفض إلى أن وصل 2017 عند حوالي 130 ألف حادث.
#
# الفايدة من الرسم:
# يوضح لنا إن فيه اتجاه عام (Trend) لانخفاض عدد الحوادث مع مرور السنوات،
# وهذا ممكن يكون نتيجة تحسينات في السلامة المرورية أو تطبيق قوانين جديدة.
```

```python
# نستخدم replace لاستبدال القيم المفقودة أو غير الصالحة
accident_df['Road_Type'] = accident_df['Road_Type'].replace("Data missing or out of range", "Unknown")
# الفايدة: نستبدل القيمة "Data missing or out of range" بالقيمة "Unknown" عشان نقدر نتعامل مع البيانات بشكل أفضل ونحسن جودتها
```

```python
# نجهز جدول يوضح عدد الحوادث لكل (نوع طريق × خطورة حادث)
road_severity = pd.crosstab(accident_df['Road_Type'], accident_df['Accident_Severity_Classified'])
# نستخدم crosstab لإنشاء جدول يوضح توزيع الحوادث حسب نوع الطريق ودرجة الخطورة

# نحول الأعداد إلى نسب مئوية عشان يكون أوضح
road_severity_percent = road_severity.div(road_severity.sum(1), axis=0)
# نستخدم div لحساب النسب المئوية، نقسم كل صف على مجموع الحوادث في نفس الصف عشان نحصل على النسب المئوية

# نرسم Stacked Bar Chart
road_severity_percent.plot(kind='bar', stacked=True, figsize=(10,6), colormap="tab20c")
# نرسم الرسم البياني العمودي (Stacked Bar) باستخدام النسب المئوية لكل نوع طريق وكل درجة خطورة مع ألوان محددة من colormap "tab20c"

plt.title("Accident Severity Distribution by Road Type")
# عنوان الرسم يوضح توزيع درجة الخطورة حسب نوع الطريق

plt.xlabel("Road Type")
# تسمية المحور الأفقي (نوع الطريق)

plt.ylabel("Proportion of Accidents")
# تسمية المحور العمودي (نسبة الحوادث)

plt.xticks(rotation=30)
# نلف أسماء المحور الأفقي (نوع الطريق) قليلاً لتكون أكثر وضوحًا

plt.legend(title="Accident Severity")
# نعرض مفتاح (Legend) يوضح ما تعنيه الألوان في الرسم البياني (الخطورة)

plt.show()  # نعرض الرسم البياني
```
